```python
from langchain_core.tools import tool
```

```python
# Step 1 - create a function

def multiply(a,b):
    """Multiply two numbers together."""
    return a * b
```

Method.1 Creating tool with @tool

```python
# Step 2 - add type hints

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers together."""
    return a * b
```

```python
result = multiply.invoke({"a": 3, "b": 4})
```

```python
print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
print(multiply.args_schema.model_json_schema())
```

Method.2 Creating tool with StucturedTool & Pydantic

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to multiply")
    b: int = Field(required=True, description="The second number to multiply")
```

```python
def multiply_func(a: int, b: int) -> int:
    return a * b

multiplty_tool = StructuredTool.from_function(
    name="multiply",
    description="Multiply two numbers together.",
    args_schema=MultiplyInput,
    func=multiply_func
)
```

```python
result = multiplty_tool.invoke({"a": 5, "b": 6})

print(result)
print(multiplty_tool.name)
print(multiplty_tool.description)   
print(multiplty_tool.args)
print(multiplty_tool.args_schema.model_json_schema())
```

Method.3 Creating tool with BaseTool

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
# arg chema using pydantic

class MultiplyInput(BaseModel):
    a: int = Field(required=True, description="The first number to multiply")
    b: int = Field(required=True, description="The second number to multiply")
```

```python
class multiplyTool(BaseTool):
    name: str = "multiply"
    description: str = "Multiply two numbers together."
    args_schema: Type[BaseModel] = MultiplyInput

    def _run(self, a: int, b: int) -> int:
        return a * b
```

```python
multiply_tool = multiplyTool()
```

```python
result = multiply_tool.invoke({"a": 3, "b": 4})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
print(multiply_tool.args_schema.model_json_schema())
```
